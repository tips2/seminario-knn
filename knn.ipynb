```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
from mlxtend.plotting import plot_decision_regions
```

```python
#Cria uma função para receber uma base de dados e um valor de K
#Então cria um modelo com o K passado na função e cria o grafico do resultado com as areas de decisão
def knn_comparison(data, k):
 x = data[['X','Y']].values
 y = data['class'].astype(int).values
 clf = neighbors.KNeighborsClassifier(n_neighbors=k)
 clf.fit(x, y)

 plot_decision_regions(x, y, clf=clf, legend=2)

 plt.xlabel('X')
 plt.ylabel('Y')
 plt.title('Knn com K='+ str(k))
 plt.show()
```

```python
data1 = pd.read_csv('data/ushape.csv')
for i in [1,5,20,30,40,80]:
    knn_comparison(data1, i)
```

```python
data2 = pd.read_csv('data/concertriccir2.csv')
for i in [1,5,20,30,40,60]:
    knn_comparison(data2, i)
```
